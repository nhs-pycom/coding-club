##**Error Handling**

Error handling enables us to plan for potential problems that may occur when executing our code: such as missing or incorrect data, human input errors e.tc. and ensure they are handled correctly to produce our intended output.

### Try-Except Blocks

Try-except blocks are one way to implement error handling in Python, a try-except block allows us to catch and handle exceptions that may occur during the execution of our code.

We can use built-in exceptions in Python from the table below to catch some common errors in our code.

Here are some examples of using a try-except block for standard errors that we can handle in Python...

Try running the following statement and see which error occurs

```python
print(x)
```

Now run the cell below with the print statement inside the try-except block

```python
try:
  print(x)
  #the code should fail and give us a NameError because x is not defined

except: #catch any errors that my occur (in this case a NameError)
  print('sorry X is not defined') 
```

We can also specify Python to only except and handle certain types of errors, like the ZeroDivisionError in the example below. 

The code will only break when a different type of error that we are not expecting occurs.

```python
## ZeroDivisionError

#the code in the try block will attempt to be executed and we can catch the error that may occur
try: 
    result = 1 / 0
#if we know the type of error that will occur we could use built-in exceptions 
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
    #everything in the except block will be executed in the event that an error occurs
```

```python
## AssertionError

#we can check data types or specific values by asserting an expression to check the variable
a = 56

try:
  assert type(a) == str #assert the expression that our variable is a string
  print('assertions complete')

except AssertionError:
    print("Error: data type is incorrect")
    #everything in the except block will be executed in the event that an error occurs

#note that the rest of the code in the try block (print statement) is not excecuted
#if an error occurs it skips straight to the exception block
```

## Custom Exceptions

In addition to handling built-in exceptions, we can also define custom exceptions class in our code. This allow us to print a custom error message for specific errors in our code.

Here's an example of a custom exception class:

```python
class NegativeNumberError(Exception): #initialise our class with a descriptive name inheriting attributes from the Exception class
    pass
```

In this example, we are defining a custom exception class called NegativeNumberError. This class inherits from the built-in Exception class, and does not define any additional methods or properties.

We need to use the **raise** function to call our custom exception to tell Python that our error has occured like follows:

```python
a = -1

try:
  if a < 0: 
    raise NegativeNumberError() #our error should trigger if our variable a is negative
  else:
    print('The number is inputted correctly')
except NegativeNumberError:
  print('Error: The number must be greater than 0') #print our custom error message 
```

# Try, Except, Else and Finally

We can add further structure to our try-except block by adding statements to tell Python what to do when an error does not occur.

```python
x = 5
y = 0

try:
  result = x / y
except ZeroDivisionError: #except block only runs when an error occurs
    print("Error: Cannot divide by zero ")
else: #this part of the block only runs if no errors occur
    print("All variables were entered correctly, the anser is: ", result)
finally: #this part of the block is always excecuted regardless of the error
    print('All calculations are complete please run the next part of the code')  
```

## Excercise 1

Write a Python program that prompts the user to enter two integers consecuitively and then divides the first number by the second number. 

Place the code int the following cell into a try-except block to handle the ZeroDivisionError that occurs if the second number entered by the user is zero and prompt the user to enter the numbers again.

**Note:** We want the code to loop over and prompt the user to enter integers until no errors occur

**Bonus:** Also add an exception if the input from the user is not an integer.

```python
while True:
  num1 = int(input("Enter the first number: "))
  num2 = int(input("Enter the second number: "))
  result = num1 / num2
  print(f"The result is {result}.")
  break
```

```python
## solution

while True:
    try:
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        result = num1 / num2
        print(f"The result is {result}.")
        break
    except ZeroDivisionError:
        print("Error: cannot divide by zero. Please try again.")
    except ValueError:
        print("Error: please enter integers only.")
```
