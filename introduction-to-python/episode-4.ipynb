# **9. Loops in python:**

Using loops in python allows a program to repeat a specific beahviour and re-run a line(s) of code. 

**for Loops:**

`for` loops are generally used when you want a process to repeat *for* a fixed number of times. 

The `for` loop always requires the following structure:

Untitled-7.png
The `variable_name` can be anything as long as it isn't another one of your variable names.

The `collection` is any element that can be iterated over. For example, lists, strings, tuples, dictionaries (the keys), and sets (in no particular order). These will be covered in more detial in a subsequent section. 

**Important note:** Notice how a colon always indicates that at least the next line of code will be indented.

By creating a loop you allo a program to repeat a conditional process. A loop will continue to execute as long as a certain condition is met. There are two types of key loops in python: `for` and `while` loops.

```python
#for loop in action.

total = 0                      #Line 1 

for num in [0, 1, 2, 3, 4, 5]: #Line 2
  total = total + num          #Line 3

print(total)                   #Line 4
```

**what just happened?**

In this example, `prime_number` is an arbitrarily variable  it could just as easily be `prime` or even` x`. Each time we loop, this variable references the next element in the list `primes` until none are left.

When Line 2 is evaluated, `prime_number` is set to the first element of `primes: primes[0]`. Effectively, it sets prime_`number = primes[0]` (**Important note:** index `0` is the number `2` in this list, index `1` is the number `3`. etc.). Because Line 3 is indented, it's evaluated next and prime_number (i.e., `primes[0]` equivalent to the number `2`) is displayed.

Now, we loop back to Line 2. Here, `prime_number` is set to the next element of `primes: primes[1]`. On Line 3, we display `prime_number` again.

This continues until we loop back to Line 2 and we've iterated through each element in the `primes` list. At this point, the `for` loop is done and we jump to the next line after the code block, which is Line 4.

```python
#for loop in action #2.

primes = [2, 3, 5, 7, 11]     #Line 1        

for prime_number in primes:   #Line 2
  print(prime_number)         #Line 3

print('Done!')                #Line 4
```

To start, the variable `total `is `0`.

In Line 2, the variable `num` is set to 0, which is the first element in the list.

In Line 3, `total = total + num`, so `total = 0 + 0`. 

Therefore at the end of the first loop of the code block, `total `is still equal to `0`.

We loop back to Line 2. Now, `num` is set to the second element, `1`. In Line 3, `total = total + num`, `so total = 0 + 1`.

Once the `for` loop is done, we see that `total` is `15` (`0` + `1` + `2` + `3` + `4` + `5`), as expected.

**while Loops:**

The `while` loop A while loop is generally used in Python when we don't know when the looping will stop or how many iterations the loop will require. It is simpler and cleaner than the `for` loop. The `while` loop requires the following structure: 

Untitled-8.png

As long as the conditional is true in a `while `loop, the loop will keep going forever



```python
#A while loop in action:

counter = 0                                                       # Line 1

while counter < 10:                                               # Line 2
  print('counter is currently', counter)                          # Line 3
  counter = counter + 1                                           # Line 4

print("We're out of the loop. The value of counter is", counter)
```

Here, we're creating a while loop using the `counter` variable.

On Line 1, we're starting `counter` off at `0`.

We then evaluate if `counter` is less than `10` (Line 2). Because `0` is less than `10`, Lines 3 and 4 are run.

Line 3: We simply print the current value of counter.
Line 4: We then increment the value of counter by `1` each loop. 
Once counter is equal to `10`, the loop will break, as `10` is not less than `10`.


# **10. NumPy Part 1:**

Lists in Python are a powerful tool. A list can hold any data of any type and can be composed of different data types at the same time. You can also change, add and remove elements. However, lists do not allow you to carry out operations over entire collections of values.

```python
#Why lists are not useful for the calculations of collections of values
#Try calculate BMI from the following two lists. BMI = kg/m^2 

height = [1.72, 1.84, 1.92, 2.03]
weight = [74.2, 92.3, 91.3, 86.8]

BMI = ((weight)/(height**2))
print(BMI)
```

Python cannot do calculations on lists. The solution is the python package **NumPy**, which stands for numeric python. NumPy provides an alternative to the python list a NumPy array. Unlike with lists, you can carry out calculations on NumPy arrays. 

```python
#Run the following cell to import th numpy package.
import numpy as np
```
